```python
def IMU_Set():
    
    IMU_Set = []

    IMUData = pd.read_csv(DIR_Data + Subject + '/IMU.csv')
    
    # Load Data #    
    AccX_L = Filter_BL(IMUData['AccX_L'])
    AccY_L = Filter_BL(IMUData['AccY_L'])
    AccZ_L = Filter_BL(IMUData['AccZ_L'])
    
    AccX_R = Filter_BL(IMUData['AccX_R'])
    AccY_R = Filter_BL(IMUData['AccY_R'])
    AccZ_R = Filter_BL(IMUData['AccZ_R'])
    
    AccX_W = Filter_BL(IMUData['AccX_W'])
    AccY_W = Filter_BL(IMUData['AccY_W'])
    AccZ_W = Filter_BL(IMUData['AccZ_W'])
    
    QX_L = Filter_BL(IMUData['QX_L'])
    QY_L = Filter_BL(IMUData['QY_L'])
    QZ_L = Filter_BL(IMUData['QZ_L'])
    QS_L = Filter_BL(IMUData['QS_L'])

    QX_R = Filter_BL(IMUData['QX_R'])
    QY_R = Filter_BL(IMUData['QY_R'])
    QZ_R = Filter_BL(IMUData['QZ_R'])
    QS_R = Filter_BL(IMUData['QS_R'])
    
    QX_W = Filter_BL(IMUData['QX_W'])
    QY_W = Filter_BL(IMUData['QY_W'])
    QZ_W = Filter_BL(IMUData['QZ_W'])
    QS_W = Filter_BL(IMUData['QS_W'])
    
    AngX_L = Filter_BL(IMUData['AngX_L'])
    AngY_L = Filter_BL(IMUData['AngY_L'])
    AngZ_L = Filter_BL(-IMUData['AngZ_L'])
    
    AngX_R = Filter_BL(IMUData['AngX_R'])
    AngY_R = Filter_BL(IMUData['AngY_R'])
    AngZ_R = Filter_BL(IMUData['AngZ_R'])
    
    AngX_W = Filter_BL(IMUData['AngX_W'])
    AngY_W = Filter_BL(IMUData['AngY_W'])
    AngZ_W = Filter_BL(-IMUData['AngZ_W'])
    
    
    # Convert Coordinate #
    CoorAccX_L = ((1-2*(QY_L**2+QZ_L**2))*AccX_L)+ \
                 ((2*(QX_L*QY_L-QZ_L*QX_L))*AccY_L)+ \
                 ((2*(QX_L*QZ_L+QY_L*QS_L)*AccZ_L))
    
    CoorAccX_R = ((1-2*(QY_R**2+QZ_R**2))*AccX_R)+ \
                 ((2*(QX_R*QY_R-QZ_R*QX_R))*AccY_R)+ \
                 ((2*(QX_R*QZ_R+QY_R*QS_R)*AccZ_R))
    
    CoorAccX_W = ((1-2*(QY_W**2+QZ_W**2))*AccX_W)+ \
                 ((2*(QX_W*QY_W-QZ_W*QX_W))*AccY_W)+ \
                 ((2*(QX_W*QZ_W+QY_W*QS_W)*AccZ_W))
    
    CoorAccY_L = ((2*(QX_L*QY_L+QZ_L*QS_L))*AccX_L)+ \
                 ((1-2*(QX_L**2+QZ_L**2))*AccY_L)+ \
                 ((2*(QY_L*QZ_L-QX_L*QS_L)*AccZ_L))
    
    CoorAccY_R = ((2*(QX_R*QY_R+QZ_R*QS_R))*AccX_R)+ \
                 ((1-2*(QX_R**2+QZ_R**2))*AccY_R)+ \
                 ((2*(QY_R*QZ_R-QX_R*QS_R)*AccZ_R))
    
    CoorAccY_W = ((2*(QX_W*QY_W+QZ_W*QS_W))*AccX_W)+ \
                 ((1-2*(QX_W**2+QZ_W**2))*AccY_W)+ \
                 ((2*(QY_W*QZ_W-QX_W*QS_W)*AccZ_W))
    
    CoorAccZ_L = ((2*(QX_L*QZ_L-QY_L*QS_L))*AccX_L)+ \
                 ((2*(QY_L*QZ_L+QX_L*QS_L))*AccY_L)+ \
                 ((1-2*(QX_L**2+QY_L**2))*AccY_L) - 9.8
    
    CoorAccZ_R = ((2*(QX_R*QZ_R-QY_R*QS_R))*AccX_R)+ \
                 ((2*(QY_R*QZ_R+QX_R*QS_R))*AccY_R)+ \
                 ((1-2*(QX_R**2+QY_R**2))*AccY_R) - 9.8
    
    CoorAccZ_W = ((2*(QX_W*QZ_W-QY_W*QS_W))*AccX_W)+ \
                 ((2*(QY_W*QZ_W+QX_W*QS_W))*AccY_W)+ \
                 ((1-2*(QX_W**2+QY_W**2))*AccY_W) - 9.8
    
    IMU_Set.append(CoorAccX_L)
    IMU_Set.append(CoorAccY_L)
    IMU_Set.append(CoorAccZ_L)

    IMU_Set.append(CoorAccX_R)
    IMU_Set.append(CoorAccY_R)  
    IMU_Set.append(CoorAccZ_R)
    
    IMU_Set.append(CoorAccX_W)
    IMU_Set.append(CoorAccY_W)  
    IMU_Set.append(CoorAccZ_W)
    
    IMU_Set.append(AngX_L)
    IMU_Set.append(AngY_L)
    IMU_Set.append(AngZ_L)
            
    IMU_Set.append(AngX_R)    
    IMU_Set.append(AngY_R)
    IMU_Set.append(AngZ_R)
    
    IMU_Set.append(AngX_W)    
    IMU_Set.append(AngY_W)
    IMU_Set.append(AngZ_W)
    
    return (IMU_Set)
```
