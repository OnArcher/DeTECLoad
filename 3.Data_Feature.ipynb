```python
def Data_Feature(Range_Sub, Label_Weight_Sub, Label_Posture_Sub):

    IMU_Data_Set = []
    Label_Weight_Set = []
    Label_Posture_Set = []

    for i in range(0,len(Range_Sub)):

        WeightNumber = Label_Weight_Sub[i]
        PostureNumber = Label_Posture_Sub[i]

        for j in range(0,5):

            Window = j * 128
            IMU_Data_Set_Temp = []

            for k in range(0,len(IMU_Set_Sub)):

                IMU = IMU_Set_Sub[k][Range_Sub[i][0]+Window:Range_Sub[i][0]+Window+384] # 3 secodns
                IMU_Data_Set.append(IMU)
                
            IMU_Data_Set_Temp.append(IMU)

            Label_Weight_Set_Temp = np.zeros(np.array(IMU_Data_Set_Temp).shape[0])
            Label_Weight_Set_Temp[:] = WeightNumber
            Label_Weight_Set = np.append(Label_Weight_Set, Label_Weight_Set_Temp)

            Label_Posture_Set_Temp = np.zeros(np.array(IMU_Data_Set_Temp).shape[0])
            Label_Posture_Set_Temp[:] = PostureNumber
            Label_Posture_Set = np.append(Label_Posture_Set, Label_Posture_Set_Temp)
            
    Feature_Set = Feature_Seg(IMU_Data_Set)
    
    return (Feature_Set, Label_Weight_Set, Label_Posture_Set)
```

```python
def Data_GAF(Feature_Set):
    
    Image_Size = 30
    GASF_Setting = GramianAngularField(Image_Size,method='summation')

    GAF_Set = []

    for i in range(0,len(Feature_Set)):

        GAF_Seg_Set = []

        for j in range(0,len(Feature_Set[0])):

            GAF = GASF_Setting.fit_transform(Feature_Set[i][j])

            GAF_Seg_Set.append(GAF)

        GAF_Set.append(GAF_Seg_Set)
        
    return(GAF_Set)
```
